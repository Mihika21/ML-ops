```python
import PyPDF2
import random
import nltk
from nltk import bigrams, FreqDist, ConditionalFreqDist
import os
import string

def convert_pdf_to_text(pdf_path, txt_path):
    """
    Convert a PDF file to a text file.
    
    Args:
    - pdf_path (str): Path to the PDF file.
    - txt_path (str): Path to save the text file.
    """
    with open(pdf_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        text = ''
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    
    with open(txt_path, 'w', encoding='utf-8') as txt_file:
        txt_file.write(text)

# Example usage:
pdf_file_path = r'D:\advanta\sanford_guidelines\Sanford Guide_ Penicillin G.pdf'  # Replace with the path to your PDF file
text_file_path = r'D:/advanta/sanford_guidelinesoutput.txt'  # Replace with the desired path for the output text file
convert_pdf_to_text(pdf_file_path, text_file_path)
print('done')
```

```python
import math

split_index = int(0.8 * len(actual_bigrams))  # 80% train, 20% validation
train_set = actual_bigrams[:split_index]
val_set = actual_bigrams[split_index:]

bigram_model_train = defaultdict(lambda: defaultdict(int))
for word1, word2 in train_set:
    bigram_model_train[word1][word2] += 1

# validation set with Laplace smoothing
V = len(set(word for bigram in train_set for word in bigram))
total_log_probability = 0
total_words = 0

for word1, word2 in val_set:
    # Count occurrences of bigram in training data with Laplace smoothing
    count_bigram = bigram_model_train[word1][word2] + 1
    # Count occurrences of word1 in training data
    count_word1 = sum(bigram_model_train[word].get(word2, 0) for word in bigram_model_train) + V

    # Calculate probability of bigram with Laplace smoothing
    probability_bigram = count_bigram / count_word1

    total_log_probability += math.log(probability_bigram)
    total_words += 1

perplexity = math.exp(-total_log_probability / total_words)
print("Perplexity on validation set with Laplace smoothing:", perplexity)
```

```python
import os
import string
import nltk
from nltk.util import bigrams
from nltk.tokenize import word_tokenize
from collections import defaultdict

punctuation = string.punctuation.replace('.', '')  # Retain the full stop

def is_hidden(filepath):
    return os.path.basename(filepath).startswith('.')

input_data_dir = r"D:\Pract1"
text_data = ""
for filename in os.listdir(input_data_dir):
    filepath = os.path.join(input_data_dir, filename)
    if not is_hidden(filepath):
        with open(filepath, encoding='utf-8') as infile:
            for line in infile:
                if line.strip():  # Check if line is not just whitespace
                    # Remove all punctuation except full stops
                    for char in punctuation:
                        line = line.replace(char, '')
                    text_data += line

words = nltk.word_tokenize(text_data.lower())
bi_grams = list(bigrams(words))

# Train the Bigram Model
bigram_model = defaultdict(lambda: defaultdict(int))
for word1, word2 in bi_grams:
    bigram_model[word1][word2] += 1

# Calculate perplexity for each bigram
for i, test in enumerate(bi_grams):
    perplexity = bigram_model.perplexity(test)
    print("PP({0}): {1}".format(test, perplexity))
```

```python
import math

def calculate_perplexity(bigram_model, test_set):
    N = 0
    log_probability_sum = 0

    for word1, word2 in test_set:
        # Count occurrences of bigram in training data
        count_bigram = bigram_model[word1][word2]

        # Count occurrences of word1 in training data
        count_word1 = sum(bigram_model[word].get(word2, 0) for word in bigram_model)

        # Calculate probability of bigram
        probability_bigram = count_bigram / count_word1 if count_word1 > 0 else 0

        if probability_bigram > 0:
            log_probability_sum += math.log(probability_bigram)
        else:
            log_probability_sum += float('-inf')  # Handling unseen bigrams

        N += 1

    # Calculate perplexity
    perplexity = math.exp(-log_probability_sum / N)
    return perplexity

# Example usage:
test_set = [("this", "is"), ("is", "a"), ("a", "test")]
perplexity = calculate_perplexity(bigram_model, test_set)
print("Perplexity:", perplexity)
```
